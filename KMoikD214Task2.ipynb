### D. Analysis

Once I had my data prepared for my analysis, I first set my dependent and independent variables, added my constant, and fit the data to my multiple linear regression mdel that I used - Ordinary Least Squares.

Multiple linear regression and, more specifically, Ordinary Least Squares was chosen for this task as this is one of the most relevant models used to determine which variables have the greatest impact on the sale price of homes. Multiple linear regression models can find relationships between one continuous dependent variable and 2 or more independent variables that are continuous, or that are categorical but have been re-expressed as numberic (Statistics Solutions, n.d.). Ordinary Least Squares will obtain the coefficients of the independent variables while minimizing the sum of the squared residuals. While the chosen method is useful for what we are trying to determine, it is important to remember that this method assumes there is a linear relationship between the dependent and independent variables and that the errors have a constant variance.  These assumptions may not be true in real-world scenarios (Sampaio, 2023).

Printing the summary of my initial model, the r-squared and adjusted r-squared values of 0.051 indicates that the variables in the model are only explaining approximately 5% of the variance in determining house prices.  However, most variables have a p-value smaller than 0.05, indicating they are statistically significant, but, city-encoded has a p-value of 0.98, indicating it is not statistically significant.  Finally, the model notes there is a chance of strong multicollinearity.

Due to the indication of possible multicollinearity, I obtained the VIF values of each variable to confirm any multicollinearity. All variables had a VIF values under 10, except for bed which is 12.06 and bath which is 12.5.  As bath has the highest, I removed it and re-checked the VIF of the remaining variables. Once bath was removed, all remaining variables had a VIF under 10.  

Once the multicollinearity was been addressed, I re-ran my OLS model without using the variable bath.  My r-squared and adjusted r-squared values were still 0.05, and city_encoded still had a p-value greater than 0.05 at 0.998.  The next step I took was to remove city_encoded from the model and re-run the model.  I used backwards stepwise elimination as due to the manageable amount of independent variables, it was simpler to begin with all of the variables in the model and then remove the least statistically significant ones. While this is an accepted method, it is important to note that this method can lead to false confidence in the model by underestimating the standard of errors, which can lead to possible overfitting, and it may not provide the best model (Smith, 2018).

After city_encoded was removed from the model, all remaining variables had a p-value less than 0.05, indicating there are all statistically relevant.  However, the r-squared and adjusted r-squared still indicate that these variables account for just 5% of the variance, suggesting there are other factors that affect the house price that are not included in our model.  The F-statistic of 2000 seems to indicate that this model is effective for explaining the variance in the sale price of houses (MathWorks, n.d.).

### E. Data Summary and Implications

Based on the final OLS model, the final equation is: price = -6.828e+05 + 4.398e+05 * bed + 12.5370 * acre_lot + 1.0671 * house_size + 46.0256 * zip_encoded

-6.828e+05 is the y-intercept, which is the predicted value of the house price when all other variables are equal to 0.  

The coefficient of 4.398e+05 for beds means, assuming all else remains constant, an increase of 1 bedroom increases the sale price by 4.398e+05.

Assuming all else remains constant, an increase of 1 unit for acre_lot results in an increase in price by 12.5370.

Assuming all else remains constant, an increase of 1 unit for house_size results in an increase in price by 1.0671.

Assuming all else remains constant, an increase in 1 unit for zip_encoded results in an increase in price by 46.0256.

Based on this equation, the number of bedrooms is the biggest determinant for house price compared to the other independent variables.  All remaining variables have a p-value less than 0.05, indicating they are all statistically significant.

Thus, we have successfully answered our initial research question of whether we could apply multiple linear regression to this dataset to determine the variables that have the most impact on the sale price of homes in Florida. While the Ordinary Least Squares Multiple Linear Regression method used was appropriate to determine which independent variables most affected the continuous dependent variable to answer the question of which variables most affect the sale price of homes in Florida, as stated above, it is important to remember that this method assumes there is a linear relationship between the dependent and independent variables and that the errors have a constant variance.  These assumptions may not be true in real-world scenarios (Sampaio, 2023). 

The r-squared and adjusted r-squared values of 0.051, indicating this model only explains about 5% of the variance in home sale prices in Florida.  The AIC is 4.751e+06, which is quite large, and indicates that the model may not be the best fit for the data.  The Durbin-Watson value of 1.6 is close to 2, which indicates there is most likely not severe autocorrelation but that there does appear to be some correlation between the errors, which violates the assumption that they are independent (Medium, 2021).  Thus, while my variables were determined to be statistically significant, based on the r-squared value, we must accept the null hypothesis as our model did not have an r-squared value greater than 0.5.

Residual plots for each independent variable in the final model were created. Several of the variables were unbalanced with most of the points being vertically spread around 0 on the x-axis. This confirms the model is generally accurate but indicates there could be heteroscedasticity (Qualtrics n.d.).

Additionally, the Q-Q Plot of the residuals confirms a normal distribution, which is in line with the assumption that there is a linear relationship between the dependent and independent variables, and also indicates the model is capturing the relationships between those variables well (Frost, n.d.).

My final metric of evaluation for the model was calculating the mean squared error and root mean squared error.  Both values were about the same at 0.97 and 0.98 respectively, indicating the model's predictions were off by about that amount on average. This low figure indicates the model is failure accurate in its predictions.

Ultimately, while the most statistically significant variables were identified with their coefficients, the model itself does not explain even half of the variance that affects the sale price of houses but appears to have a high rate of accuracy based on the MSE and RMSE.  It is not surprising that the variables did not explain all or even most of the variance affecting sale price as there are so many other factors not accounted for in this dataset such as recent upgrades to the home, the quality of nearby schools, amenities, environmental risks, and cost of living, all of which can vary greatly in just a small area.  My first recommendation is to obtain additional information regarding the sold houses that would potentially affect the sale price. Additionally, the final model still indicates there is possible high multicollinearity, so another variable selection tool should be explored such as PCA or ridge regression to attempt reduce any multicollinearity that is still present.  Finally, it may be beneficial to explore other model options such as Random Forests, which may handle the nulls and outliers better, and is less prone to overfitting while also being able to capture more complex relationships that may exist between the variables and the sale price of homes.
